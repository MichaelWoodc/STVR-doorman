```python
import json
import os

input_path = "media/geography_files/Parcel_Digest_2025.geojson"
output_dir = "media/geography_files/split_parcels"
os.makedirs(output_dir, exist_ok=True)

# Target chunk size in MB (aiming for ~30MB per file to stay well under GitHub limits)
TARGET_CHUNK_MB = 30
target_bytes = TARGET_CHUNK_MB * 1024 * 1024

print(f"Loading and splitting: {input_path}...")

with open(input_path, "r", encoding="utf-8") as f:
  data = json.load(f)

features = data.get("features", [])
print(f"Total features to process: {len(features)}")

chunk_index = 1
current_features = []
current_size = 0

for i, feature in enumerate(features):
  # Estimate feature size by dumping to string
  feature_str = json.dumps(feature)
  feature_size = len(feature_str.encode("utf-8"))

  if current_size + feature_size > target_bytes and current_features:
    # Save current chunk
    chunk_data = {"type": "FeatureCollection", "features": current_features}
    chunk_filename = os.path.join(
        output_dir, f"Parcel_Digest_2025_part{chunk_index}.geojson"
    )

    with open(chunk_filename, "w", encoding="utf-8") as out:
      json.dump(chunk_data, out)

    print(
        f"Saved {chunk_filename} ({len(current_features)} features,"
        f" {current_size / (1024*1024):.2f} MB)"
    )

    chunk_index += 1
    current_features = []
    current_size = 0

  current_features.append(feature)
  current_size += feature_size

# Save any remaining features in the last chunk
if current_features:
  chunk_data = {"type": "FeatureCollection", "features": current_features}
  chunk_filename = os.path.join(
      output_dir, f"Parcel_Digest_2025_part{chunk_index}.geojson"
  )
  with open(chunk_filename, "w", encoding="utf-8") as out:
    json.dump(chunk_data, out)
  print(
      f"Saved {chunk_filename} ({len(current_features)} features,"
      f" {current_size / (1024*1024):.2f} MB)"
  )

print("Splitting complete! All parts saved in:", output_dir)
```
