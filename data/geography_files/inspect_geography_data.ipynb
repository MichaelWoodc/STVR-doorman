```python
import json

# Path to your local GeoJSON parcel file
geojson_path = "media/geography_files/Parcel_Digest_2025.geojson"

print(f"Inspecting: {geojson_path}\n")

with open(geojson_path, "r", encoding="utf-8") as f:
  # Using ijson or just loading line-by-line if it's massive,
  # but standard json.load works if memory allows.
  data = json.load(f)

print(f"GeoJSON Type: {data.get('type')}")
features = data.get("features", [])
print(f"Total Parcels Found: {len(features)}\n")

if features:
  print("--- Sample Parcel Properties (First Feature) ---")
  sample_props = features[0].get("properties", {})
  for key, value in sample_props.items():
    print(f"  {key}: {value}")
else:
  print("No features found in the GeoJSON file.")
```
